```python
NER = spacy.load("en_core_web_sm")
```

## Load Books

Source https://github.com/dworschak/Witcher/tree/master/RESSOURCES/_books/text

```python
all_books =  [b for b in os.scandir('data') if '.txt' in b.name]
all_books
```

```python
# prototype using one book

book = all_books[1]
book_text = open(book).read()
book_doc = NER(book_text)
```

```python
# visualize the identified entities in the first 2000 words in the book
displacy.render(book_doc[:2000], style='ent', jupyter=True)
```
